# 2. Feature Engineering

### 2.1. Создание новых признаков

Для определения общей мощности потоков сложим значения "Силы Левого Потока" и "Силы Правого Потока", чтобы получить представление о совокупной мощности.

```python
df['Совокупная мощность'] = df['Сила Левого Потока'] + df['Сила Правого Потока']
```

Для определения суммарной силы всех потоков объединим "Силу Левого Потока", "Силу Правого Потока" и "Приток Силы Потока".

```python
df['Суммарная сила потоков'] = df['Сила Левого Потока'] + df['Сила Правого Потока'] + df['Приток Силы Потока']
```

Общую силу ядра оценим, умножив "Ритм магического ядра" на "Приток Силы Потока", что даст показатель общей мощности ядра.

```python
df['Общая сила ядра'] = df['Ритм магического ядра'] * df['Приток Силы Потока']
```

Общее давление на выходе рассчитаем, сложив значения "Притока давления Выдоха Истока" и "Давления выдоха Истока".

```python
df['Общее давление на выходе'] = df['Приток давления Выдоха Истока'] + df['Давление выдоха Истока']
```

Магическую производительность определяет отношение "Скорости перехода через портал" к "Эмульсии Истока" - это покажет, насколько эффективно топливо поддерживает скорость перехода.

```python
df['Магическая производительность'] = df['Скорость перехода через портал'] / df['Эмульсия Истока']
```

Эффективность самого ядра вычислим, разделив его общую силу на "Эмульсию Истока".

```python
df['Эффективность ядра'] = df['Общая сила ядра'] / df['Эмульсия Истока']
```

Для расчёта магической мощности учтем "Эмульсию Истока", общее давление и разницу между "Пламенем Стихий" и "Температурой вдоха Истока".\
Произведение этих величин показывает, что для достижения высокой магической мощности нужны все три компонента одновременно. Это похоже на физические процессы, где результат зависит от совместного действия нескольких факторов (например, тепловая мощность в физике зависит от массы, температуры и теплоёмкости). Если какой-либо из этих компонентов мал, магическая мощность будет снижаться, даже если другие высоки.

```python
df['Магическая мощность'] = df['Эмульсия Истока'] * (df['Пламя Стихий'] - df['Температура вдоха Истока']) * df['Общее давление на выходе']
```

Для решения поставленной задачи в условии было предложено создать еще 3 новых признака над имеющимися, содержащие целевую переменную. Это признаки: **Степень износа магических источников**, **Расхождения в стабильности магии** и **Баланс угасания**. Я принял решение не идти данным путем, поскольку явное использование значений целевой переменной в признаковом пространстве хоть и повышает линейную зависимость с таргетом, однако с точки зрения поставленной задачи это бессмысленный шаг, который приведет к переобучению модели на данной выборке и не позволит спрогнозировать значения **Гармонии Бессмертия** на новых входных данных.

```python
df.insert(len(df.columns)-1, 'Гармония Бессмертия', df.pop('Гармония Бессмертия'))
```

### 2.2. Метод гланых компонент ($PCA$) и линейная регрессия

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

X = df.drop(columns=['Гармония Бессмертия'])
y = df['Гармония Бессмертия']

# Нормализация признаков
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Алгоритм подбора числа компонент
results = []
max_components = min(X_scaled.shape[1], 20)
cv_folds = 10

for n_components in range(2, max_components + 1):
    pca = PCA(n_components=n_components)
    X_pca = pca.fit_transform(X_scaled)

    # Разделение на обучающую и тестовую выборки
    X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

    model = LinearRegression()

    # Обучение и оценка на обучающей и тестовой выборках
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    r2_difference = abs(train_r2 - test_r2)

    # Кросс-валидация для оценки устойчивости
    mse_cv = cross_val_score(model, X_pca, y, cv=cv_folds, scoring='neg_mean_squared_error')
    r2_cv = cross_val_score(model, X_pca, y, cv=cv_folds, scoring='r2')

    # Сохранение метрик
    results.append({
        'Number of Components': n_components,
        'Train MSE': train_mse,
        'Test MSE': test_mse,
        'Train R^2': train_r2,
        'Test R^2': test_r2,
        'R^2 Difference': r2_difference,
        'Mean MSE (CV)': -mse_cv.mean(),
        'Mean R^2 (CV)': r2_cv.mean(),
        'R^2 Variance (CV)': r2_cv.var()
    })

# График накопленной объяснённой дисперсии
pca = PCA(n_components=max_components)
pca.fit(X_scaled)
explained_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(8, 5))
plt.plot(range(2, max_components + 2), explained_variance_ratio, marker='o')
plt.xlabel('Число главных компонент')
plt.ylabel('Накопленная объясненная дисперсия')
plt.title('Накопленная объясненная дисперсия от числа главных компонент')
plt.grid(True)
plt.show()

# Создание DataFrame для удобного анализа
results_df = pd.DataFrame(results).sort_values(by='Number of Components', ascending=False)
results_df['Cumulative Explained Variance (%)'] = sorted(explained_variance_ratio[:max_components-1], reverse=True)
display(results_df)

plt.figure(figsize=(8, 5))
plt.plot(range(2, max_components + 1), results_df.sort_values(by='Number of Components', ascending=True)['R^2 Difference'], marker='o')
plt.xlabel('Число главных компонент')
plt.ylabel('R^2 Difference')
plt.title('Абсолютная разница R^2 на тестовой и обучающей выборках от числа главных компонент')
plt.grid(True)
plt.show()
```

В качестве оптимального значения числа главных компонент обычно выбирается порог в 95% объясненной дисперсии данных. С целью определить оптимальное значение, предотвратив переобучение модели, был построен график абсолютной разницы R^2 на обучающей и тестовой выборках. Предполагаем, что если такая разница в метриках при увеличении числа главных компонент возрастает, то модель переобучается.

Таким образом, оптимальное число главных компонент: 7

```python
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV

optimal_pca = PCA(n_components=7)
X_pca = optimal_pca.fit_transform(X_scaled)

r2_cv = cross_val_score(model, X_pca, y, cv=cv_folds, scoring='r2')

# Добавим регуляризатор Elastic Net к линейной регрессии с помощью GridSearchCV для подбора гиперпараметров
elastic_net = ElasticNet()
param_grid = {
    'alpha': [0.01, 0.1, 0.5, 1.0, 5.0, 10.0, 50.0, 100.0],  # Расширенный диапазон для регуляризации
    'l1_ratio': [0.1, 0.3, 0.5, 0.7, 0.9, 1.0]               # Различные значения соотношения L1 и L2
}
grid_search = GridSearchCV(elastic_net, param_grid, cv=10, scoring='r2')
grid_search.fit(X_pca, y)

# Обучение Elastic Net с оптимальными параметрами
best_elastic_net = grid_search.best_estimator_
best_elastic_net.fit(X_pca, y)
print(f'Mean R^2 LinearRegression (CV): {r2_cv.mean()}')
print(f'Mean R^2 ElasticNet (CV): {grid_search.best_score_}')
```

**Вывод:** Простая множественная линейная регрессия после устранения мультиколлинеарности методом главных компонент ($PCA$) плохо описывает закономерности в наборе данных, с регуляризацией результат ее работы становится только хуже.

```python
# Получение матрицы компонентов
components =  optimal_pca.components_

# Преобразование в DataFrame для удобства
# Имена столбцов - это исходные признаки, а строки - главные компоненты
components_df = pd.DataFrame(components, columns=X.columns)

plt.figure(figsize=(10, 6))
sns.heatmap(components_df, cmap="coolwarm", annot=True)
plt.xlabel("Исходные признаки")
plt.ylabel("Главные компоненты (PC)")
plt.title("Тепловая карта PCA - вклад исходных признаков")
plt.show()
```

### 3.1. Полиномиальная регрессия

Попробуем обучить полиномиальную регрессию, предполагая, что комбинации признаков, а точнее их произведения, подобно признаку **Магическая производительность**, лучше опишут зависимости с целевой переменной. Число новых признаков с увеличением степени регрессии увеличивается комбинаторно.

#### Число признаков = $\sum_{k=1}^{p} \binom{d + k - 1}{k}$, где в сумме бином-е коэф-ты $\binom{n}{k} = \frac{n!}{k!(n - k)!}$, d - число исходных признаков


* 1 степень: 7 признаков (главных компонент)
* 2 степень: 7 + 28 признаков
* 3 степень: 7 + 28 + 84 признаков \

Сначала обучим линейную регрессию на главных компонентах и их комбинациях 2 и 3 степеней.

```python
# Для построения графиков и интерпретации результатов модели
from scipy.stats import probplot
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.stats.stattools import durbin_watson

def diagnostics(y, y_pred):
  residuals = y - y_pred
  residuals_mean = np.round(np.mean(y - y_pred), 3)

  f, ((ax_rkde, ax_prob), (ax_ry, ax_auto), (ax_yy, ax_ykde)) = plt.subplots(nrows = 3,
                                                                             ncols = 2,
                                                                             figsize = (6, 9))

  # в первом подграфике построим график плотности остатков
  sns.kdeplot(residuals, fill = True, ax = ax_rkde)
  ax_rkde.set_title('Residuals distribution', fontsize = 10)
  ax_rkde.set(xlabel = f'Residuals, mean: {residuals_mean}')
  ax_rkde.set(ylabel = 'Density')

  # во втором - график нормальной вероятности остатков
  probplot(residuals, dist = 'norm', plot = ax_prob)
  ax_prob.set_title('Residuals probability plot', fontsize = 10)

  # в третьем - график остатков относительно прогноза
  ax_ry.scatter(y_pred, residuals)
  ax_ry.set_title('Predicted vs. Residuals', fontsize = 10)
  ax_ry.set(xlabel = 'y_pred')
  ax_ry.set(ylabel = 'Residuals')

  # в четвертом - автокорреляцию остатков
  plot_acf(residuals, lags = 30, ax = ax_auto)
  ax_auto.set_title('Residuals Autocorrelation', fontsize = 10)
  ax_auto.set(xlabel = f'Lags \ndurbin_watson: {durbin_watson(residuals).round(2)}')
  ax_auto.set(ylabel = 'Autocorrelation')

  # в пятом - сравним прогнозные и фактические значения
  ax_yy.scatter(y, y_pred)
  ax_yy.plot([y.min(), y.max()], [y.min(), y.max()], "k--", lw = 1)
  ax_yy.set_title('Actual vs. Predicted', fontsize = 10)
  ax_yy.set(xlabel = 'y_true')
  ax_yy.set(ylabel = 'y_pred')

  # в шестом - сравним распределение истинной и прогнозной целевой переменных
  sns.kdeplot(y, fill = True, ax = ax_ykde, label = 'y_true')
  sns.kdeplot(y_pred, fill = True, ax = ax_ykde, label = 'y_pred')
  ax_ykde.set_title('Actual vs. Predicted Distribution', fontsize = 10)
  ax_ykde.set(xlabel = 'y_true and y_pred')
  ax_ykde.set(ylabel = 'Density')
  ax_ykde.legend(loc = 'upper right', prop = {'size': 6})

  plt.tight_layout()

  plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

pca = PCA(n_components=7)
X_pca = pca.fit_transform(X_scaled)

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
y_val = y_test

# Создание полиномиальных признаков второй степени
poly_features = PolynomialFeatures(degree=2)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)
X_val_poly = X_test_poly

# Линейная регрессия на полиномиальных признаках
model = LinearRegression()

# Кросс-валидация на обучающей выборке (10 фолдов)
cv_scores = cross_val_score(model, X_train_poly, y_train, cv=10, scoring='r2')
print(f"Среднее значение R^2 на обучающей выборке (10-кратная кросс-валидация): {cv_scores.mean()}")

# Обучение модели на обучающей выборке
model.fit(X_train_poly, y_train)

# Предсказание на тестовой выборке
y_pred = model.predict(X_test_poly)

# Оценка качества модели на тестовой выборке
test_r2 = r2_score(y_test, y_pred)
print(f"R^2 на тестовой выборке: {test_r2}")

diagnostics(y_test, y_pred)
```

```python
def plot_learning_curves(model, X_train_poly, y_train, X_val_poly, y_val):
  train_errors, val_errors = [], []
  for m in range(1, len(X_train_poly), 100):
    model.fit(X_train_poly[:m], y_train[:m])
    y_train_predict = model.predict(X_train_poly[:m])
    y_val_predict = model.predict(X_val_poly)
    train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
    val_errors.append(mean_squared_error(y_val, y_val_predict))
  plt.plot(np.sqrt(train_errors), "r+", linewidth=2, label="train")
  plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="val")
  plt.xlabel("Размер обучающего набора")
  plt.ylabel("RMSE")
  plt.title("Кривая обучения модели")
  plt.legend()
  plt.show()

plot_learning_curves(model, X_train_poly, y_train, X_val_poly, y_val)
```

### 4.1. Дерево решений ($DesicionTree$)

Это одиночное дерево, которое строится на основе разделений данных по признакам с целью максимизации чистоты узлов. Оно следует иерархической структуре, где каждый узел делит данные на две части до тех пор, пока не будут достигнуты листья, которые представляют классы (для классификации) или значения (для регрессии). Модель предоставляет важность признаков, что помогает понять, какие параметры оказывают наибольшее влияние на целевую переменную.

```python
from sklearn.tree import DecisionTreeRegressor

# Разделение на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Инициализация модели дерева решений
dt_model = DecisionTreeRegressor(max_depth=13, random_state=42)

# Кросс-валидация на обучающей выборке
cv_scores = cross_val_score(dt_model, X_train, y_train, cv=5, scoring='r2')
cv_mean = cv_scores.mean()
cv_std = cv_scores.std()

# Обучение модели на всей обучающей выборке
dt_model.fit(X_train, y_train)

# Прогнозирование на обучающей выборке для оценки метрик
y_pred_train = dt_model.predict(X_train)
mse_train = mean_squared_error(y_train, y_pred_train)
r2_train = r2_score(y_train, y_pred_train)

# Прогнозирование на тестовой выборке для оценки метрик
y_pred_test = dt_model.predict(X_test)
mse_test = mean_squared_error(y_test, y_pred_test)
r2_test = r2_score(y_test, y_pred_test)

# Вывод результатов
print("Cross-validation R^2 Mean:", cv_mean)
print("Cross-validation R^2 Std Dev:", cv_std)
print("Test MSE:", mse_test)
print("Test R^2:", r2_test)
```

### 4.2. Случайный лес ($RandomForest$)

Случайный Лес хорошо справляется с мультиколлинеарностью и нелинейными зависимостями между признаками, благодаря усреднению предсказаний нескольких деревьев, устойчив к шуму и менее подвержен переобучению по сравнению с отдельными деревьями решений.
